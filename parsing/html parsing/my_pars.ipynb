```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re
# импортируем библиотеки
```

```python
url = 'https://topnaroda.com/rating/kino/37161-top-100-filmov-vseh-vremen.html' # переменная, содержащая в себе ссылку на сайт для парсинга
```

```python
pages = requests.get(url) # подключаемся
soup = bs(pages.text, 'html.parser') # определеяем
print(pages.status_code) # вывод статуса состояния
```

```python
result_list = {'title_ru': [], 'title_eng': [], 'year': [], 'genre': [], 'duration': [], 'country': [], 'director': [], 'quote':[], 'main_actors':[], 'rating_KinoPoisk':[], 'rating_IMDB':[], 'description':[]}
```

```python
%time
# этот цикл для того, чтобы перемещаться по страницапм
# Пустые переменные для записи туда значений
title_ru = ''
title_eng = ''
year = 0
genre = ''
duration = ''
country = ''
director = ''
description = ''
quote = ''
main_actors = ''
rating_KinoPoisk = ''
rating_IMDB = ''

# юрлка
url = 'https://topnaroda.com/rating/kino/37161-top-100-filmov-vseh-vremen.html' # тут видно что переходим на страницу
print(url) # вывожу чтобы видеть процесс парсинга

kino = requests.get(url)
soup = bs(kino.text, 'html.parser')

data = soup.find('div', class_='grid clearfix grid-list')

if data is None:
    print('data is None!!!')

titletext = data.find_all('div', class_='s-right5')
ratings = data.find_all('div', class_='frates clearfix')

for i in titletext:
    urldesc = str(i.a.get('href')) # юрлка

    desc = requests.get(urldesc) # Подключаюсь
    soup2 = bs(desc.text, 'html.parser')
    try:
        description = soup2.find('div', style="font-size: 16px; line-height:23px;").text
    except:
        description = '-'
    print(description)

    span_name = soup2.find('span', string="Название:")
    try:
        title_eng = span_name.next_sibling.strip()
    except AttributeError:
        title_eng = '-'
    print(title_eng)

    span_genre = soup2.find('span', string="Жанр:")
    try:
        genre = span_genre.next_sibling.strip()
    except AttributeError:
        genre = '-'
    print(genre)

    span_country = soup2.find('span', string="Страна:")
    try:
        country = span_country.next_sibling.strip()
    except AttributeError:
        country = '-'
    print(country)

    span_director = soup2.find('span', string="Режиссер:")
    try:
        director = span_director.next_sibling.strip()
    except AttributeError:
        director = '-'
    print(director)

    span_quote = soup2.find('span', string="Слоган:")
    try:
        quote = span_quote.next_sibling.strip()
    except AttributeError:
        quote = '-'
    print(quote)

    span_time = soup2.find('span', string="Время:")
    try:
        duration = span_time.next_sibling.strip()
    except AttributeError:
        duration ='-'
    print(duration)

    span_actors = soup2.find('span', string="В ролях:")
    try:
        actors = span_actors.next_sibling.strip().split(',')[0:5]
        main_actors = ','.join(actors)
    except:
        main_actors = '-'
    print(main_actors)

    year = int(i.find('a', class_='s-title').text.split(' ')[-1].replace(')', '').replace('(',''))    
    title_ru = i.find('a', class_='s-title').text.split('(')[0].strip()
    
    
    result_list['description'].append(description)
    result_list['genre'].append(genre)
    result_list['title_eng'].append(title_eng)
    result_list['title_ru'].append(title_ru)
    result_list['country'].append(country)
    result_list['director'].append(director)
    result_list['quote'].append(quote)
    result_list['duration'].append(duration)
    result_list['main_actors'].append(main_actors)
    result_list['year'].append(year)

for i in ratings:
    try:
        span_rating1 = i.find('span', string=' КиноПоиск - ')
        if span_rating1:
            r1 = span_rating1.next_sibling.strip()
            if r1 == '':
                r1 = 0.0
    except:
        r1 = 0.0
            
    try:
        span_rating2 = i.find('span', string='IMDB - ')
        if span_rating2:
            r2 = span_rating2.next_sibling.strip()
            if r2 == '':
                r2 = 0.0  
    except:
        r2 = 0.0
        
    print(r1)
    print(r2)
            
    result_list['rating_KinoPoisk'].append(r1)
    result_list['rating_IMDB'].append(r2)
```

```python
# Просто вывод длины для проверки
print(len(result_list['title_ru']))
print(len(result_list['title_eng']))
print(len(result_list['year']))
print(len(result_list['country']))
print(len(result_list['genre']))
print(len(result_list['director']))
print(len(result_list['duration']))
print(len(result_list['description']))
print(len(result_list['quote']))
print(len(result_list['main_actors']))
print(len(result_list['rating_KinoPoisk']))
print(len(result_list['rating_IMDB']))
```
