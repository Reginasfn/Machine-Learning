## Извлечение текста из PDF

```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

## Парсинг текста из PDF.

```python
def extract_publication_date(text):    
    # регулярные для поиска определенного формата даты
    date_regex = re.compile(r'(\d{1,2} [а-я]+ в \d{1,2}:\d{2})')  # 28 фев в 13:40
    date_year = re.compile(r'(\d{1,2} [а-я]+ \d{4} в \d{1,2}:\d{2})')  # 28 ноя 2024 в 16:01
    time_ago_regex = re.compile(r'(\d+) час(?:а|ов)? назад')  # 6 часов назад или 1 час назад
    time_min = re.compile(r'(\d+) минут(?:у|ы)? назад') # 12 минут назад
    # поиск даты публикации
    for line in text:
        date_match = date_regex.search(line) #ищем в строках похожее
        if date_match:
            return date_match.group(1) #находим по регулярному выраж
        
        date_year_match = date_year.search(line)
        if date_year_match:
            return date_year_match.group(1)
            
        time_ago_match = time_ago_regex.search(line)
        if time_ago_match:
            num = int(time_ago_match.group(1)) #учитываем разные окончания
            
            if num == '1' or num == '21':
                return f"{num} час назад"
            elif 2 <= num <= 4 or 22 <= num <= 24:
                return f"{num} часa назад"
            elif 4 <= num <= 20:
                return f"{num} часов назад"
            else:
                return line
        
        time_min_match = time_min.search(line)
        if time_min_match:
            num = int(time_min_match.group(1))

            if 1 <= num <= 59:  #в диапазоне от 1 до 59
                if num == 1:
                    return f"{num} минуту назад"
                elif 2 <= num <= 4:
                    return f"{num} минуты назад"
                else:
                    return f"{num} минут назад"
            else:
                return f"{num} минут назад"

    return None  #если дата не найдена!!!


def search_activity(text2):
                   
    time_regex = re.compile(f'(\d+) {re.escape("мин")}')  #содержит число за которым слово мин
    for line in text2:
        match = time_regex.search(line)
        if match:
            time = match.group(1)
            try:
                
                activity_i = text2.index(line) + 5 #индекс строки с акктивити
                activity = text2[activity_i]
                
                i = activity_i + 1 #6
                while i < len(text2): #пока не превышает размера текста
                    if text2[i].strip() == ',' and i + 1 < len(text2): # если запятая и есть в границах текста
                        activity = activity + ', ' + text2[i + 1].strip().lower()  # добавляем пока след элемент запятая
                        i += 2
                    else:
                        break
                        
                        
                if not text2[text2.index(line) + 5].startswith('Блог'): #исключения
                    activity = text2[text2.index(line) + 2]
                    
                return activity
            except (ValueError, IndexError):
                return text2[text2.index(line) + 2]
    return None



def pars_pdf(text):
    text2 = text.split('\n')
    text2 = [elem.strip().strip('*').replace('\xa0', ' ') for elem in text2 if elem.strip()] #лишние пробелы, символы
    print(text2)
    
    date_publish = ''
    if 'Рейтинг' in text2: #отталкиваемся от рейтинга
        num = text2.index('Рейтинг')
        
        raiting = text2[num - 1]
        description = '-'
        
        if num + 2 < len(text2) and not text2[num + 2]: #в пределах и не пустая
            description = '-'
        else:
            check = text2[num + 2]
            if check != 'Подписаться' and len(check) <= 50: 
                description = check
            else:
                description = '-'
        
        name_comp = text2[num + 1]
        
        activity = search_activity(text2) #поиск активити
        date_publish = extract_publication_date(text2) #поиск даты
         
        print(num, name_comp, description, raiting, date_publish, activity)
        
    else: #обработка исключительных файлов
        description = "-"
        raiting = 0.0
        name_comp = "-"
        activity = "-"
        
        date_publish = extract_publication_date(text2).split(' ')
        print(date_publish)
        if not date_publish[0].isdigit():
            name_comp = date_publish[0]
        else:
            try:
                i = text2.index(extract_publication_date(text2))  # Ищем '5' в text2
                name_comp = text2[i + 2]  # Название компании находится перед '5'
            except ValueError:
                name_comp = "-"
                
        if len(date_publish) == 4:
            date_publish = ' '.join(date_publish[1:])
        else:
            date_publish = extract_publication_date(text2)
        
        print(name_comp, description, raiting, date_publish, activity)
    
    try:
        raiting = float(raiting)  # Пробуем преобразовать в float
    except (ValueError, TypeError):  # Если преобразование не удалось
        return None
        
    return name_comp, description, raiting, date_publish, activity
```
