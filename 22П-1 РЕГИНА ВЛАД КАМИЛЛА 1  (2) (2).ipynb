```python
from sklearn.preprocessing import LabelEncoder
```

```python
type_data = list(set(sh['Eye color'].values))
labelencoder = LabelEncoder()
sh['Eye color'] = labelencoder.fit_transform(sh['Eye color'])
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(sh["Eye color"])
```

Оценили график выбросов и сделали вывод, что есть герои с редким цветом глаз

```python
type_data = list(set(sh['Gender'].values))
labelencoder = LabelEncoder()
sh['Gender'] = labelencoder.fit_transform(sh['Gender'])

type_data = list(set(sh['Race'].values))
labelencoder = LabelEncoder()
sh['Race'] = labelencoder.fit_transform(sh['Race'])

type_data = list(set(sh['Hair color'].values))
labelencoder = LabelEncoder()
sh['Hair color'] = labelencoder.fit_transform(sh['Hair color'])

type_data = list(set(sh['Publisher'].values))
labelencoder = LabelEncoder()
sh['Publisher'] = labelencoder.fit_transform(sh['Publisher'])

type_data = list(set(sh['Publisher'].values))
labelencoder = LabelEncoder()
sh['Publisher'] = labelencoder.fit_transform(sh['Publisher'])

type_data = list(set(sh['Skin color'].values))
labelencoder = LabelEncoder()
sh['Skin color'] = labelencoder.fit_transform(sh['Skin color'])

type_data = list(set(sh['Alignment'].values))
labelencoder = LabelEncoder()
sh['Alignment'] = labelencoder.fit_transform(sh['Alignment'])
```

Он преобразует текстовые категории (такие как пол, раса и цвет волос и т.д.) в числовые значения, что позволяет использовать эти данные в алгоритмах машинного обучения, которые требуют числовых входов.

```python
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(sh[['Gender', 'Eye color', 'Race', 'Hair color', 'Height', 'Publisher', 'Skin color', 'Alignment', 'Weight']].corr(), annot=True, linewidths=.5, ax=ax, cmap='gist_earth',fmt = '0.2f', alpha=.6)
plt.show()
```

Посторили тепловую карту, показывающую корреляцию между выбранными столбцами. Что позволяет нам визуально оценить, как сильно связаны различные переменные между собой.
